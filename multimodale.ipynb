```python
import json
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
import os
```

```python
with open("image_text_embeddings.json", "r", encoding="utf-8") as f:
    image_text_embeddings = json.load(f)

with open("image_visual_embeddings.json", "r", encoding="utf-8") as f:
    image_visual_embeddings = json.load(f)

print(f"Images avec embeddings visuels: {len(image_visual_embeddings)}")
print(f"Images avec embeddings textuels: {len(image_text_embeddings)}")
```

On va venir créer un dico pour les embeddings

```python
multimodal_embeddings = {}
```

Soit on concatenate, soit on fait la moyenne

```python
for image_path, vis_emb in image_visual_embeddings.items():
    if image_path in image_text_embeddings:
        text_emb = image_text_embeddings[image_path]

        # Convertir les listes en tableaux numpy
        vis_emb_np = np.array(vis_emb)
        text_emb_np = np.array(text_emb)
        combined_emb = np.concatenate((vis_emb_np, text_emb_np))
        multimodal_embeddings[image_path] = combined_emb.tolist() # Stocker comme liste pour JSON
    else:
        print(f"Avertissement: Pas d'embedding textuel trouvé pour {image_path}. L'image sera ignorée pour la recherche multimodale.")

print(f"Embeddings multimodaux combinés pour {len(multimodal_embeddings)} images.")
```

```python
for image_path, vis_emb in image_visual_embeddings.items():
    if image_path in image_text_embeddings:
        text_emb = image_text_embeddings[image_path]

        # Convertir les listes en tableaux numpy
        vis_emb_np = np.array(vis_emb)
        text_emb_np = np.array(text_emb)

        # --- Stratégies de combinaison ---

        # Stratégie 1: Concaténation (la plus simple)
        # Assurez-vous que les embeddings sont bien des vecteurs 1D
        combined_emb = np.concatenate((vis_emb_np, text_emb_np))

        # Stratégie 2: Somme ou moyenne (si les dimensions sont similaires)
        # Si vous voulez faire une somme ou moyenne, il faudrait que les dimensions soient les mêmes
        # Ou alors, utiliser des techniques comme la projection.
        # Pour l'instant, on se concentre sur la concaténation.

        multimodal_embeddings[image_path] = combined_emb.tolist() # Stocker comme liste pour JSON
    else:
        print(f"Avertissement: Pas d'embedding textuel trouvé pour {image_path}. L'image sera ignorée pour la recherche multimodale.")

print(f"Embeddings multimodaux combinés pour {len(multimodal_embeddings)} images.")
```

```python
output_path_multimodal = "image_multimodal_embeddings.json"
with open(output_path_multimodal, "w", encoding="utf-8") as f:
    json.dump(multimodal_embeddings, f)
print(f"Embeddings multimodaux sauvegardés dans {output_path_multimodal}")
```
